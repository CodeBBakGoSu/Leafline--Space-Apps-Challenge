```python
import pandas as pd

# ---------------------------
# 1️⃣ 기본 필터 유지
# ---------------------------
species_filter = ['red maple', 'eastern redbud']
phenophase_filter = ['Flowers or flower buds', 'Open flowers', 'Pollen release (flowers)']
state_filter = 'FL'

def filter_df(df, need_phenophase=True):
    cond = df['Common_Name'].isin(species_filter) & (df['State'] == state_filter)
    if need_phenophase and 'Phenophase_Description' in df.columns:
        cond &= df['Phenophase_Description'].isin(phenophase_filter)
    return df.loc[cond]

df_site = filter_df(df_site, need_phenophase=False)
df_mag = filter_df(df_mag)
df_state = filter_df(df_state)
df_ind = filter_df(df_ind)

# ---------------------------
# 2️⃣ df_mag을 연도별 요약으로 축소
# ---------------------------
agg_cols = ['Common_Name', 'Phenophase_Description']
numeric_cols = df_mag.select_dtypes(include='number').columns.tolist()

df_mag_summary = df_mag.groupby(agg_cols)[numeric_cols].mean().reset_index()

# ---------------------------
# 3️⃣ df_state + df_ind 먼저 병합
# ---------------------------
df_state_ind = pd.merge(
    df_state,
    df_ind,
    on=['Common_Name', 'Phenophase_Description', 'State', 'Latitude', 'Longitude'],
    how='outer',
    suffixes=('_state', '_ind')
)

# ---------------------------
# 4️⃣ df_site 추가 병합 (좌표+이름 기준)
# ---------------------------
merged_site = pd.merge(
    df_state_ind,
    df_site,
    on=['Common_Name', 'State', 'Latitude', 'Longitude'],
    how='left',
    suffixes=('', '_site')
)

# ---------------------------
# 5️⃣ 마지막으로 df_mag_summary를 조인
# ---------------------------
final_df = pd.merge(
    merged_site,
    df_mag_summary,
    on=['Common_Name', 'Phenophase_Description'],
    how='left',
    suffixes=('', '_mag')
)

# ---------------------------
# 6️⃣ 정리
# ---------------------------
final_df.drop_duplicates(subset=['Common_Name', 'Phenophase_Description', 'Latitude', 'Longitude'], inplace=True)
final_df.reset_index(drop=True, inplace=True)

print(f"✅ 병합 완료: {final_df.shape}")
```
