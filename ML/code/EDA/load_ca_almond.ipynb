1. 개화 직전 (Pre-flowering)

'Budburst' (싹이 틈 → 생육 시작)

'First flower bud' (첫 꽃봉오리)

'Flower buds' (꽃봉오리들)
👉 개화가 임박했음을 알리는 신호 단계입니다. 기온·일조량과 함께 개화 예측의 중요한 입력값.

2. 개화기 (Flowering stage)

'First flower' (첫 꽃)

'First bloom' (첫 개화)

'Open flowers' (꽃이 핀 상태)

'Full bloom' (만개)

'Full flowering' (전면 개화)
👉 실제 개화 예측의 핵심 태그로, 모델이 가장 집중해야 할 구간.

3. 개화 종료 (End of flowering)

'End bloom' (개화 종료)

'End of flowering' (꽃이 다 진 시점)

👉 꽃이 진 이후 시기로, 전체 생육/열매 맺음 단계로 넘어가는 기준점.

🌾 4. 보조적 태그 (수분·꽃가루 관련)

'First pollen released' (첫 꽃가루 방출)

'Full pollen release' (꽃가루 전량 방출)

'Pollen release' (꽃가루 방출)

'Open pollen cones' (개방된 수꽃송이 → 침엽수에 중요)

👉 개화와 밀접하게 연계되지만, “꽃가루 시기”를 직접 예측하는 모델에서 더 유용.
다만 개화 모델에서도 추가 피처로 활용 가능.
