```python
import rasterio
from rasterio.plot import show
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pyproj import Proj, Transformer

# 1. ÌååÏùº Ïó¥Í∏∞
# tif_file = "your_file_path_here.tif"  # Ïã§Ï†ú ÌååÏùº Í≤ΩÎ°úÎ°ú ÏàòÏ†ïÌïòÏÑ∏Ïöî

with rasterio.open(path) as src:
    evi_raw = src.read(1).astype(np.float32)
    transform = src.transform
    crs = src.crs
    width = src.width
    height = src.height

# 2. EVI Ïä§ÏºÄÏùºÎßÅ (Ïä§ÏºÄÏùº Ìå©ÌÑ∞ 10000)
evi_scaled = evi_raw / 10000.0

# 3. Í≤∞Ï∏°Ïπò Ï≤òÎ¶¨ (nodataÍ∞Ä Î™ÖÏãúÏ†ÅÏúºÎ°ú ÏóÜÏßÄÎßå ÏùºÎ∞òÏ†ÅÏúºÎ°ú -3000 Îì±ÏúºÎ°ú Ï†ïÏùòÎêòÍ∏∞ÎèÑ Ìï®)
evi_scaled = np.where((evi_scaled < -1.0) | (evi_scaled > 1.0), np.nan, evi_scaled)

# 4. Í∏∞Î≥∏ ÌÜµÍ≥Ñ Ï∂úÎ†•
print("üìä Í∏∞Î≥∏ ÌÜµÍ≥Ñ")
print(f"ÏµúÏÜå: {np.nanmin(evi_scaled)}")
print(f"ÏµúÎåÄ: {np.nanmax(evi_scaled)}")
print(f"ÌèâÍ∑†: {np.nanmean(evi_scaled)}")
print(f"ÌëúÏ§ÄÌé∏Ï∞®: {np.nanstd(evi_scaled)}")

# 5. ÌûàÏä§ÌÜ†Í∑∏Îû®
plt.figure(figsize=(8, 4))
sns.histplot(evi_scaled[~np.isnan(evi_scaled)].flatten(), bins=100, kde=True)
plt.title("EVI Î∂ÑÌè¨ (Ïä§ÏºÄÏùº Ï†ÅÏö© ÌõÑ)")
plt.xlabel("EVI")
plt.ylabel("ÌîΩÏÖÄ Ïàò")
plt.grid(True)
plt.tight_layout()
plt.show()

# 6. ÏãúÍ∞ÅÌôî (ÏßÄÎèÑ)
plt.figure(figsize=(12, 6))
plt.imshow(evi_scaled, cmap='YlGn', vmin=0, vmax=1)
plt.colorbar(label='EVI')
plt.title("EVI Ïä§ÏºÄÏùºÎêú ÏßÄÎèÑ ÏãúÍ∞ÅÌôî")
plt.axis('off')
plt.show()

# 7. ÏúÑÎèÑ/Í≤ΩÎèÑ Ï∂îÏ∂ú Ìï®Ïàò
def pixel_to_latlon(row, col, transform, src_crs):
    """Ìñâ, Ïó¥ -> ÏúÑÎèÑ Í≤ΩÎèÑ"""
    # (x, y) Ï¢åÌëúÍ≥Ñ ÏúÑÏπò
    x, y = rasterio.transform.xy(transform, row, col, offset='center')

    # sinusoidal ‚Üí WGS84 Î≥ÄÌôò
    transformer = Transformer.from_crs(src_crs, "EPSG:4326", always_xy=True)
    lon, lat = transformer.transform(x, y)
    return lat, lon

# ÏòàÏãú: Ï§ëÏïô ÏßÄÏ†êÏùò ÏúÑÎèÑ/Í≤ΩÎèÑ Ï∂îÏ∂ú
center_row = height // 2
center_col = width // 2
lat, lon = pixel_to_latlon(center_row, center_col, transform, crs)
print(f"üåê Ï§ëÏïô ÏßÄÏ†ê ÏúÑÎèÑ/Í≤ΩÎèÑ: ({lat:.6f}, {lon:.6f})")

# ÏòàÏãú: EVI Í∞íÏù¥ ÎÜíÏùÄ ÏßÄÏ†êÏùò ÏúÑÎèÑ/Í≤ΩÎèÑ Ï∂îÏ∂ú
max_idx = np.nanargmax(evi_scaled)
max_row, max_col = np.unravel_index(max_idx, evi_scaled.shape)
lat_max, lon_max = pixel_to_latlon(max_row, max_col, transform, crs)
print(f"üìçÏµúÍ≥† EVI ÏßÄÏ†ê: ({lat_max:.6f}, {lon_max:.6f}), EVI: {evi_scaled[max_row, max_col]:.4f}")
```

```python
import glob
import rasterio
from rasterio.merge import merge
from rasterio.warp import calculate_default_transform, reproject, Resampling
import matplotlib.pyplot as plt
import numpy as np
import os

# 1. TIF ÌååÏùº Î™©Î°ù Î∂àÎü¨Ïò§Í∏∞
tif_dir = r"C:\ÏÑ∏Ìúò\NASAÌï¥Ïª§ÌÜ§\data\EARTHDATA\order2"  # Ìè¥Îçî Í≤ΩÎ°ú ÏßÄÏ†ï
tif_files = glob.glob(os.path.join(tif_dir, "*.tif"))

# 2. Î≥ëÌï©
src_files_to_mosaic = [rasterio.open(fp) for fp in tif_files]
mosaic, out_trans = merge(src_files_to_mosaic)

# 3. Î≥ëÌï©Îêú Î©îÌÉÄÎç∞Ïù¥ÌÑ∞ Í∞±Ïã†
out_meta = src_files_to_mosaic[0].meta.copy()
out_meta.update({
    "height": mosaic.shape[1],
    "width": mosaic.shape[2],
    "transform": out_trans,
    "count": 1
})

# 4. Ï¢åÌëúÍ≥Ñ Ïû¨Ìà¨ÏòÅ (Sinusoidal ‚Üí WGS84)
dst_crs = 'EPSG:4326'
transform, width, height = calculate_default_transform(
    src_files_to_mosaic[0].crs, dst_crs,
    mosaic.shape[2], mosaic.shape[1], *src_files_to_mosaic[0].bounds)

# Ïû¨Ìà¨ÏòÅ ÎåÄÏÉÅ Î∞∞Ïó¥
reprojected = np.empty((height, width), dtype=np.float32)

reproject(
    source=mosaic[0],
    destination=reprojected,
    src_transform=out_trans,
    src_crs=src_files_to_mosaic[0].crs,
    dst_transform=transform,
    dst_crs=dst_crs,
    resampling=Resampling.nearest
)

# 5. ÏãúÍ∞ÅÌôî (ÏÑ∏Í≥ÑÏßÄÎèÑ Ï¢åÌëú Í∏∞Ï§Ä)
plt.figure(figsize=(12, 6))
plt.imshow(reprojected, cmap='YlGn', extent=[-180, 180, -90, 90])
plt.title("Ï†Ñ ÏßÄÍµ¨ EVI (WGS84Î°ú Î≥ÄÌôò)")
plt.colorbar(label="EVI")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.tight_layout()
plt.show()
```
