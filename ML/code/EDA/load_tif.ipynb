```python
import rasterio
from rasterio.plot import show
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pyproj import Proj, Transformer

# 1. 파일 열기
# tif_file = "your_file_path_here.tif"  # 실제 파일 경로로 수정하세요

with rasterio.open(path) as src:
    evi_raw = src.read(1).astype(np.float32)
    transform = src.transform
    crs = src.crs
    width = src.width
    height = src.height

# 2. EVI 스케일링 (스케일 팩터 10000)
evi_scaled = evi_raw / 10000.0

# 3. 결측치 처리 (nodata가 명시적으로 없지만 일반적으로 -3000 등으로 정의되기도 함)
evi_scaled = np.where((evi_scaled < -1.0) | (evi_scaled > 1.0), np.nan, evi_scaled)

# 4. 기본 통계 출력
print("📊 기본 통계")
print(f"최소: {np.nanmin(evi_scaled)}")
print(f"최대: {np.nanmax(evi_scaled)}")
print(f"평균: {np.nanmean(evi_scaled)}")
print(f"표준편차: {np.nanstd(evi_scaled)}")

# 5. 히스토그램
plt.figure(figsize=(8, 4))
sns.histplot(evi_scaled[~np.isnan(evi_scaled)].flatten(), bins=100, kde=True)
plt.title("EVI 분포 (스케일 적용 후)")
plt.xlabel("EVI")
plt.ylabel("픽셀 수")
plt.grid(True)
plt.tight_layout()
plt.show()

# 6. 시각화 (지도)
plt.figure(figsize=(12, 6))
plt.imshow(evi_scaled, cmap='YlGn', vmin=0, vmax=1)
plt.colorbar(label='EVI')
plt.title("EVI 스케일된 지도 시각화")
plt.axis('off')
plt.show()

# 7. 위도/경도 추출 함수
def pixel_to_latlon(row, col, transform, src_crs):
    """행, 열 -> 위도 경도"""
    # (x, y) 좌표계 위치
    x, y = rasterio.transform.xy(transform, row, col, offset='center')

    # sinusoidal → WGS84 변환
    transformer = Transformer.from_crs(src_crs, "EPSG:4326", always_xy=True)
    lon, lat = transformer.transform(x, y)
    return lat, lon

# 예시: 중앙 지점의 위도/경도 추출
center_row = height // 2
center_col = width // 2
lat, lon = pixel_to_latlon(center_row, center_col, transform, crs)
print(f"🌐 중앙 지점 위도/경도: ({lat:.6f}, {lon:.6f})")

# 예시: EVI 값이 높은 지점의 위도/경도 추출
max_idx = np.nanargmax(evi_scaled)
max_row, max_col = np.unravel_index(max_idx, evi_scaled.shape)
lat_max, lon_max = pixel_to_latlon(max_row, max_col, transform, crs)
print(f"📍최고 EVI 지점: ({lat_max:.6f}, {lon_max:.6f}), EVI: {evi_scaled[max_row, max_col]:.4f}")
```

```python
import glob
import rasterio
from rasterio.merge import merge
from rasterio.warp import calculate_default_transform, reproject, Resampling
import matplotlib.pyplot as plt
import numpy as np
import os

# 1. TIF 파일 목록 불러오기
tif_dir = r"C:\세휘\NASA해커톤\data\EARTHDATA\order2"  # 폴더 경로 지정
tif_files = glob.glob(os.path.join(tif_dir, "*.tif"))

# 2. 병합
src_files_to_mosaic = [rasterio.open(fp) for fp in tif_files]
mosaic, out_trans = merge(src_files_to_mosaic)

# 3. 병합된 메타데이터 갱신
out_meta = src_files_to_mosaic[0].meta.copy()
out_meta.update({
    "height": mosaic.shape[1],
    "width": mosaic.shape[2],
    "transform": out_trans,
    "count": 1
})

# 4. 좌표계 재투영 (Sinusoidal → WGS84)
dst_crs = 'EPSG:4326'
transform, width, height = calculate_default_transform(
    src_files_to_mosaic[0].crs, dst_crs,
    mosaic.shape[2], mosaic.shape[1], *src_files_to_mosaic[0].bounds)

# 재투영 대상 배열
reprojected = np.empty((height, width), dtype=np.float32)

reproject(
    source=mosaic[0],
    destination=reprojected,
    src_transform=out_trans,
    src_crs=src_files_to_mosaic[0].crs,
    dst_transform=transform,
    dst_crs=dst_crs,
    resampling=Resampling.nearest
)

# 5. 시각화 (세계지도 좌표 기준)
plt.figure(figsize=(12, 6))
plt.imshow(reprojected, cmap='YlGn', extent=[-180, 180, -90, 90])
plt.title("전 지구 EVI (WGS84로 변환)")
plt.colorbar(label="EVI")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from pyproj import Transformer

# Sinusoidal (MODIS) → WGS84 변환기
src_crs = src.crs  # Sinusoidal
dst_crs = "EPSG:4326"  # WGS84

transformer = Transformer.from_crs(src_crs, dst_crs, always_xy=True)
lon, lat = transformer.transform(x, y)
print("위도/경도:", lat, lon)
```

```python
from rasterio.warp import transform_bounds

bounds = src.bounds  # (xmin, ymin, xmax, ymax)
bounds_wgs84 = transform_bounds(src.crs, "EPSG:4326", *bounds)
print("WGS84 경계:", bounds_wgs84)
# WGS84 경계: (서쪽 경도, 남쪽 위도, 동쪽 경도, 북쪽 위도)
```
